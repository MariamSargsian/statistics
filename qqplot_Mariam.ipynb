```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats
```

```python
def qqplot(np_dist, dist='norm'):
    x = np.linspace(0,1,len(np_dist))
    plt.figure(figsize=(10,10))
    np_dist = np.sort(np_dist)
    plt.xlabel("Theoretical Quantiles")
    plt.ylabel("Sample Quantiles")
    if dist =='binom':
        res = st.binom.ppf(x, 10, 0.5)
    else:
        res = st.norm.ppf(x, loc = 0, scale = 1)
    #plt.plot(n,n,color='blue')
    plt.plot(res,np_dist,'bo')
```

```python
a = np.random.normal(10, 7, 1000)
qqplot(a)
```

```python
bin = np.random.binomial(7, 0.3, 1000)
qqplot(bin)
```

```python
uni  = np.random.uniform(low=10, high=1000, size=1000)
qqplot(uni)
```

```python
np_normal = pd.Series(np.random.normal(0, 1, 200))
sm.qqplot(np_normal,line='45',fit=True,dist=stats.norm)
```

```python
from scipy.stats import skewnorm

skewness = -5 
random = skewnorm.rvs(a = skewness,loc=1000, size=50000)  #Skewnorm function

random = random - min(random)     
random = random / max(random)      
random = random * 50000  
random = pd.Series(random)
sm.qqplot(random,fit=True,line='45') 
plt.show()
```
