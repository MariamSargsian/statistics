```python
#3
import numpy as np
import matplotlib.pyplot as plt
def foo(n):
    dist = []
    for i in range(1000):
        sample = np.random.lognormal(3,1,n)
        dist.append(np.mean(sample))
    plt.hist(dist,alpha=0.2)
```

```python
foo(10)
```

```python
foo(20)
```

```python
foo(30)
```

```python
foo(50)
```

```python
#2
for n in [5,10,20,30,40,500]:
    iqr = []
    for i in range(100):
        x = np.random.uniform(8,10,n)
        q1, q3 = np.percentile(x,[25, 75])
        iqr.append(q3 - q1)
    print(f"for sample size {n} distribution mean is  {np.mean(iqr)} and variance {np.var(iqr)}")
```

```python
#5
corr = []
for n in range(10000):
    x = np.random.uniform(0,1,20)
    y = np.random.uniform(0,1,20)
    correlation = np.corrcoef(x, y)[0][1]
    corr.append(correlation)
plt.hist(corr)
```

```python
import pylab as py
import scipy.stats as stats
```

```python
stats.probplot(corr, dist="norm", plot=py)
py.show()
```

```python
cor = np.array(corr)
a = np.where((cor>-0.3) & (cor<0.3))
prob = cor[a]
p = len(prob)/len(cor)
print(p)
```
